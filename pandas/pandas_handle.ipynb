```python
import numpy as np
```

```python
url = (
     "https://raw.github.com/pandas-dev"
     "/pandas/master/pandas/tests/io/data/csv/tips.csv"
   )

tips = pd.read_csv(url)
```

```python
tips.head(5)
```

```python
# SELECT total_bill, tip, smoker, time
# FROM tips;
# in pandas
tips[['total_bill','tip','smoker','time']]
```

```python
# SELECT *, tip/total_bill as tip_rate
# FROM tips;
tips.assign(tip_rate = tips['tip'] / tips['total_bill'])
```

```python
# SELECT *
# FROM tips
# WHERE time = 'Dinner';
tips[tips["total_bill"] > 10]
```

```python
# SELECT *
# FROM tips
# WHERE time = 'Dinner' AND tip > 5.00;
tips[(tips['time'] == 'Dinner') & (tips['tip'] > 5.00)]
```

```python
# SELECT *
# FROM tips
# WHERE size >= 5 OR total_bill > 45;
tips[(tips['size'] > 5 ) | (tips['total_bill'] > 45)]
```

```python
# SELECT sex, count(*)
# FROM tips
# GROUP BY sex;
tips.groupby('sex').size()
```

```python
tips.groupby('sex').count()
```

```python
tips.groupby("sex")["total_bill"].count()
```

```python
# SELECT day, AVG(tip), COUNT(*)
# FROM tips
# GROUP BY day;
# /*
tips.groupby('day').agg({'tip':np.mean, 'day':np.size})
```

```python
# SELECT smoker, day, COUNT(*), AVG(tip)
# FROM tips
# GROUP BY smoker, day;
tips.groupby(['smoker','day']).agg({'tip':[np.size,np.mean]})
```

```python
df1 = pd.DataFrame({"key": ["A", "B", "C", "D"], "value": np.random.randn(4)})
df2 = pd.DataFrame({"key": ["B", "D", "D", "E"], "value": np.random.randn(4)})
df2
```

```python
# SELECT *
# FROM df1
# INNER JOIN df2
#   ON df1.key = df2.key;
# merge() also offers parameters for cases when you’d like to join one DataFrame’s column with another
pd.merge(df1,df2, on='key')
```

```python
# SELECT *
# FROM df1
# LEFT OUTER JOIN df2
#   ON df1.key = df2.key;
pd.merge(df1,df2, on='key', how='left')
```

```python
# SELECT *
# FROM df1
# RIGHT OUTER JOIN df2
#   ON df1.key = df2.key;
pd.merge(df1,df2, on='key', how='right')
```

```python
#  FULL JOINs, which display both sides of the dataset, whether or not the joined columns find a match. As of writing, FULL JOINs are not supported in all RDBMS (MySQL)
pd.merge(df1, df2, on="key", how="outer")
```

```python
df1 = pd.DataFrame({"city": ["Chicago", "San Francisco", "New York City"], "rank": range(1, 4)})
df2 = pd.DataFrame({"city": ["Chicago", "Boston", "Los Angeles"], "rank": [1, 4, 5]})
#    SELECT city, rank
# FROM df1
# UNION ALL
# SELECT city, rank
# FROM df2;
pd.concat([df1,df2])
```

```python
# SQL’s UNION is similar to UNION ALL, however UNION will remove duplicate rows.
pd.concat([df1,df2]).drop_duplicates()
```
